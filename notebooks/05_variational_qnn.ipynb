# Variational Quantum Neural Network (VQNN)

This notebook trains a VQNN for anomaly detection.

```python
import torch
import sys
import os

# Add project root to path for imports
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from src.data.load_data import load_config, load_creditcard_data
from src.data.preprocess import preprocess_creditcard
from src.features.dimensionality_reduction import apply_pca
from src.quantum.variational_qnn import VQNNTrainer

config = load_config()
```

```python
df = load_creditcard_data(config)
df = preprocess_creditcard(df, config)
X = df.drop(columns=['Class']).values
y = df['Class'].values

# PCA
n_qubits = config['quantum']['n_qubits']
X_pca, _ = apply_pca(X, n_components=n_qubits)

# Split (Simple wrapper or from src.split)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, shuffle=True)
```

```python
trainer = VQNNTrainer(n_qubits=n_qubits, n_layers=config['quantum']['n_layers'])
trainer.fit(X_train, y_train)
```
